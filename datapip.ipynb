```python
import pandas as pd
```

```python
data1=pd.read_csv("PG_DB_results/UNIPROT_sym/Sym_UniProtDB.all.tab", sep="\t", header=None)
data2=pd.read_csv("PG_DB_results/UNIPROT_sym/UniProt_BioGRID.tab", sep="\t")
data3=pd.read_csv("PG_DB_results/UNIPROT_sym/UniProt_ChEMBL.tab", sep="\t")
#data4=pd.read_csv("PG_DB_results/UNIPROT_sym/UniProt_ComplexPortal.tab", sep="\t")
#data5=pd.read_csv("PG_DB_results/UNIPROT_sym/UniProt_DIP.tab", sep="\t")
#data6=pd.read_csv("PG_DB_results/UNIPROT_sym/UniProt_DrugBank.tab", sep="\t")
data7=pd.read_csv("PG_DB_results/UNIPROT_sym/UniProt_GOID.tab", sep="\t")
data8=pd.read_csv("PG_DB_results/UNIPROT_sym/UniProt_STRING.tab", sep="\t")
```

```python
data1= data1.rename(columns={0: "GeneSymbol", 1:"Uniprot"})
```

```python
data1
```

```python
result=data1
for data in [data2,data3,data7,data8]:# [data2,data3,data4,data5,data6,data7,data8]:
    data=data.rename(columns={"Entry":"Uniprot"})
    result=pd.merge(result,data,on="Uniprot", how="left")
```

```python
result
```

```python
result=result.drop(columns= [ 2,  'UniProt_BioGRID',  'UniProt_ChEMBL', 'UniProt_DB',  'UniProt_STRING']) # add  'UniProt_ComplexPortal',  'UniProt_DIP', 'UniProt_DrugBank', 
```

```python
result=result.rename(columns={ 'Cross-reference (BioGRID)': 'BioGRID', 'Cross-reference (ChEMBL)': 'ChEMBL', 'Cross-reference (ComplexPortal)': 'ComplexPortal', 'Cross-reference (DIP)': 'DIP', 'Cross-reference (DrugBank)': 'DrugBank', 'Cross-reference (STRING)': 'STRING'})
```

```python
result
```

```python
def importer(elem,name):
    data=pd.read_csv("PG_DB_results/HGNC_sym/Sym_{}.tab".format(elem), sep="\t", header=None).rename(columns={0:"GeneSymbol",1:name,2:"drop"}).drop(columns=["drop"])
    return data
```

```python
data9=importer('EnsemblID','EnsemblID')
#data10=importer('EnzID.2','EnzID')
data11=importer('HGNC_ID.all', 'HGNC')
data12=importer('Name', 'Name')
data13=importer('NCBI_id','NCBI')
data14=importer('PrevSym_SymAlias.final', 'Alias')
```

```python
for data in[data9,data11,data12,data13,data14]:# [data9,data10,data11,data12,data13,data14]:
    result=pd.merge(result,data,on="GeneSymbol", how="left")
```

```python
result = result[~result['GeneSymbol'].str.contains(';')]
```

```python
result['BioGRID']=result['BioGRID'].astype(str).str.split('.').str[0]
result['NCBI']=result['NCBI'].astype(str).str.split('.').str[0]
```

```python
result['HGNC']=result['HGNC'].str.split(':').str[1]
```

```python
result.to_csv("datafull2.csv")
```
